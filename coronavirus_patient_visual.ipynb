```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# import data
patient_data = pd.read_csv("./data/patient.csv")
```

```python
# 0 female, 1 male
patient_data.groupby('sex')['id'].nunique().plot(kind='bar')
plt.show()
```

```python
min_year = patient_data['birth_year'].min()
max_year = patient_data['birth_year'].max()
num_bins = int(max_year-min_year)
ax2 = patient_data['birth_year'].plot.hist(bins=num_bins//5, alpha=0.5)
```

```python
patient_data.groupby('birth_year')['id'].nunique().nlargest(5).plot(kind='bar')
plt.show()
```

```python
patient_data.groupby('region')['id'].nunique().nlargest(5).plot(kind='bar')
plt.show()
```

```python
patient_data.groupby('infection_reason')['id'].nunique().plot(kind='bar')
plt.show()
```

```python
patient_data.groupby('state')['id'].nunique().plot(kind='bar')
plt.show()
```

```python
patient_data.groupby('infection_order')['id'].nunique().plot(kind='bar')
plt.show()
```

```python
patient_data['confirmed_date'] = pd.to_datetime(patient_data['confirmed_date'])
patient_data['released_date'] = pd.to_datetime(patient_data['released_date'])
patient_data['isolated_days'] = patient_data['released_date'] - patient_data['confirmed_date']
patient_data.groupby('isolated_days')['id'].nunique().plot(kind='bar')
plt.show()
```
